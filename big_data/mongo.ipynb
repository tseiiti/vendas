```python
# executar usando shell

from unidecode import unidecode
from faker import Faker
from django.contrib.auth.models import User

fake = Faker('pt_BR')
for i in range(10):
  first_name = fake.first_name()
  last_name = fake.last_name()
  email = fake.email()
  username = unidecode(first_name.split()[0].lower()) + '.'
  username += unidecode(last_name.split()[-1].lower())
  user = User.objects.create_user(
    username=username, 
    email=email, 
    password='123', 
    first_name = first_name, 
    last_name = last_name,
  )
  
```

```python
import sqlite3
con = sqlite3.connect("../db.sqlite3")
cur = con.cursor()

cur.execute("select * from auth_user where first_name <> '';")
for t in cur.fetchall():
  print(t[10], t[5])
```

```python
import sqlite3
con = sqlite3.connect("../db.sqlite3")
cur = con.cursor()

representantes = []
cur.execute("select * from auth_user where first_name <> '';")
i = 0
for t in cur.fetchall():
  representante = {
    'id': i + 1,
    'nome': f'{t[10]} {t[5]}',
    'cpf': fake.cpf(),
    'user_id': t[0]
  }
  i += 1
  representantes.append(representante)
representantes
```

```python
clientes = []
for i in range(50):
  cliente = {
    'id': i + 1,
    'nome': f'{fake.company().upper()} {fake.company_suffix().upper()}',
    'cnpj': fake.cnpj(),
    'endereco': fake.street_address(),
    'bairro': fake.bairro(),
    'cidade': fake.city(),
    'estado': fake.estado_sigla(),
    'cep': fake.postcode(formatted = True),
    'contato': fake.name(),
    'telefone': fake.cellphone_number(),
    'email': fake.email(),
    'representante_id': random.randrange(1, 21)
  }
  clientes.append(cliente)
clientes
```

# MongoDB

```python
from pymongo import MongoClient
try:
  uri = "mongodb://root:root@localhost/"
  client = MongoClient(uri)
  client.admin.command("ping")
  db = client["vendas"]
except Exception as e:
  print(e)
  exit()

# print(client.list_database_names())
# print(db.collection_names())
```

```python
db["venda_representante"].delete_many({})
db["venda_cliente"].delete_many({})
```

```python
db["venda_representante"].insert_many(representantes)
db["venda_cliente"].insert_many(clientes)
```
